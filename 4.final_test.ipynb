```python
import pandas as pd
import csv
import time
```

```python
#Filtered from PER_calculation, 2020MVP.csv contains all 2019-2020 season games 
#active players every game with PER calculations   
MVP2020_df = pd.read_csv("2020MVP.csv")
MVP2020_df.head()
```

```python
MVP2020_df['TEAM'] = MVP2020_df['MATCHUP'].str[:3]
MVP2020_df.head()
```

```python
PER20 = MVP2020_df[['TEAM','player', 'Player_ID', 'MATCHUP','WL', 'PTS','REB', 'AST','PLUS_MINUS', 'PER']]
PER20.head()
```

```python
PER20['PER'] = PER20['PER'].astype(float)
PER20_df = PER20.loc[PER20['PER'] >= 15]
PER2020 = PER20_df.loc[PER20_df['PER'] <= 40]
```

```python
PER2020.columns
```

```python
PER2020 = PER20.groupby(['player', 'TEAM'])
PER2020_df = PER2020[['PLUS_MINUS','PER']].mean()
PER2020_df.nlargest(20, ['PLUS_MINUS', 'PER'])
```
