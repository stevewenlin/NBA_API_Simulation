```python
import pandas as pd
import csv
import time
from PER import PER_function #refer to PER.py
```

```python
from nba_api.stats.static import players
player_dict = players.get_active_players()#all active players 

#making a player_id + name DF from NBA_API in order to merge other DFs 
player_dict_df = pd.DataFrame(player_dict)
full_name_id = player_dict_df[['id','full_name']]
renamed_full_name_id = full_name_id.rename(columns={'id': 'Player_ID', 'full_name': 'player'})
renamed_full_name_id.head()
```

```python
player_df = pd.read_csv('alldatabasketball.csv')
player_df = player_df.dropna(how='any')

player2_df = player_df[['Player_ID', 'Game_ID', 'GAME_DATE', 'MATCHUP', 'WL','MIN', 'FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT', 
                        'FTM', 'FTA','FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'PF','PTS', 'PLUS_MINUS']]

player2_df.head() #this DF contains all data from 2002-present, WITH NO FULL NAMES 
```

```python
player3_df = pd.merge(player2_df,renamed_full_name_id, on='Player_ID')
player3_df = player3_df.set_index("player")
player3_df.head() #this DF has Full Names now + PlayerID 
```

```python
with open ('FinalData.csv', 'a') as f: 
          player3_df.to_csv(f,mode='a', header=f.tell()==0) 
```
