```python
#this is the first step, after installing nba_api api client,
#there are  "static" modules to access player/team info without having to submit request 

from nba_api.stats.static import players
player_dict = players.get_active_players() 
player_dict #all active players 
```

```python
player_id = [player['id'] for player in player_dict] #shows all unique player IDs
player_id #this will serve as the unique identifier for all active players 
```

```python
from nba_api.stats.endpoints import playergamelog

log_player_per_season = playergamelog.PlayerGameLog(player_id='2544', season='2019') #2544 Lebron in 2019-2020 season
log_player_per_season_df = log_player_per_season.get_data_frames()[0]
log_player_per_season_df.head() #log_player_per_season_df retrieves individual stats for THAT season
```

```python
from nba_api.stats.static import teams  #now its time to get info on LAKERS & the BUCKS 

nba_teams = teams.get_teams()
lakers = [team for team in nba_teams if team['abbreviation'] == 'LAL'][0]
lakers_id = lakers['id']
```

```python
from nba_api.stats.endpoints import leaguegamefinder #all this could be found on documentation/sample page 

gamefinder = leaguegamefinder.LeagueGameFinder(team_id_nullable=lakers_id)
games = gamefinder.get_data_frames()[0]
games.head()
```

```python
games_1920 = games[games.SEASON_ID.str[-4:] == '2019']
with open ('lakers.csv', 'a') as f: 
            games_1920.to_csv(f,mode='a', header=f.tell()==0) 
```

```python
nba_teams = teams.get_teams()
bucks = [team for team in nba_teams if team['abbreviation'] == 'MIL'][0]
bucks_id = bucks['id']
```

```python
gamefinder = leaguegamefinder.LeagueGameFinder(team_id_nullable=bucks_id)
games2 = gamefinder.get_data_frames()[0]
games2.head()
```

```python
games2_1920 = games2[games2.SEASON_ID.str[-4:] == '2019']
with open ('bucks.csv', 'a') as f: 
            games2_1920.to_csv(f,mode='a', header=f.tell()==0) 
```
